```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
df['school_state'] = encode.fit_transform(df['school_state'])
df['teacher_prefix'] = encode.fit_transform(df['teacher_prefix'])
df['project_grade_category'] = encode.fit_transform(df['project_grade_category'])
df['clean_categories'] = encode.fit_transform(df['clean_categories'])
df['clean_subcategories'] = encode.fit_transform(df['clean_subcategories'])
```

```python
df.head()
```

```python
df['essay'] = df['essay'].fillna('')
df['essay'] = df['essay'].astype(str)

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['essay']).toarray()
tfidf_df = pd.DataFrame(tfidf_matrix, columns=vectorizer.get_feature_names_out())
df = pd.concat([df, tfidf_df], axis=1)
```

```python
x = df.drop(['project_is_approved'],axis=1)
y = df['project_is_approved']
```

```python
# from sklearn.compose import make_column_transformer
# from sklearn.preprocessing import OneHotEncoder
# col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['school_state','teacher_prefix','project_grade_category','clean_categories','clean_subcategories']),remainder='passthrough')
# x=col_trans.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler(with_mean=False)
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, classification_report
print(accuracy_score(y_test,y_pred))
print(roc_auc_score(y_test,y_pred))
matrix = confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(matrix, annot=True, fmt='d')
plt.show()
```

```python
import numpy as np
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
for f in range(x_train.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
# hyperparameter tuning
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [100,200,300],
    'max_depth': [5,10,15],
    'min_samples_split': [2,5,10]}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid,scoring='r2', cv=5, n_jobs=-1)
```

```python
grid_search.fit(x_train, y_train)
```
